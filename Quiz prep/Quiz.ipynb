```python
import pandas as pd 
import numpy as np
```

```python
df = pd.read_csv('bikeshare.csv')
```

```python
df.head()
```

```python
df.dropna(subset = ['start_station_id'], inplace = True)
```

```python
df['start_station_id'].isna().sum()
```

```python
df['member_gender'].nunique()
```

```python
df['member_gender'].isna().sum()
df['member_gender'] = df['member_gender'].fillna('Other')
```

```python
median_birth_year = df['member_birth_year'].median()
df['member_birth_year'] = df['member_birth_year'].fillna(median_birth_year)
df.head()
```

```python
##How many trips have a duration of less than 5 min (300 seconds?)
df.loc[df['trip_duration_sec'] < 300]
```

```python
##How many rides have been taken from start station id 15 to end station id 30? 
start15 = df.loc[df['start_station_id'] == 15] 

start15.loc[start15['end_station_id'] == 30]
```

```python
##Compute the busiest route, and find how many rides have been in the busiest route 
df2 = df.groupby(['start_station_id', 'end_station_id']).sort_values(ascending = False)
```
